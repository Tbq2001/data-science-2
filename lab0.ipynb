```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
np.random.seed(101)
```

```python
rand_matrix = np.random.randint(1,101,(100,5))
print(rand_matrix)
```

```python
plt.imshow(rand_matrix, aspect=0.05)
plt.colorbar()
plt.title('Matrix Plot')
plt.show()
```

```python
matrix_df = pd.DataFrame(data=rand_matrix)
print(matrix_df)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_matrix = scaler.fit_transform(rand_matrix)
print(scaled_matrix)
```

```python
matrix_df.columns = ['f1','f2','f3','f4','label']
print(matrix_df)
```

```python
from sklearn.model_selection import train_test_split
X = matrix_df[['f1','f2','f3','f4']]
y = matrix_df['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
print("X_train:\n", X_train)
print("X_test:\n", X_test)
print("y_train:\n", y_train)
print("y_test:\n", y_test)
```
